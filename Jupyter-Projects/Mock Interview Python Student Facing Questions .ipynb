<b> Q4. Please create two bar plots as per below that show:
    
    1) The number of individuals who have a High School Graduate Diploma AND earn <=50K in the United States
    2) The number of individuals who have a High School Graduate Diploma AND earn >50K in the United States 

Please note you will be looking specifically at the *Job Role* column

<b> Put Your Code Below </b>

```python
#filter1
#filter individuals who who have a High School Graduate Diploma AND earn <=50K in the United States
dtf1 = dtf_filter[ (dtf_filter["Degree Status"] == " HS-grad") & (dtf_filter["Earnings"] == " <=50K") 
            & (dtf_filter["Country"] == " United-States")]
#filter2
#filter individuals who have a High School Graduate Diploma AND earn >50K in the United States 
dtf2 = dtf_filter[ (dtf_filter["Degree Status"] == " HS-grad") & (dtf_filter["Earnings"] == " >50K") 
            & (dtf_filter["Country"] == " United-States")]

#to count the individuals of each Job Role, we subset the Job-Role column and use the value_counts method
#we use the sort argument to sort the Job Role (default True)
#we use the ascending argument to get the Job Role with the biggest count on top (default False)
s1_jobrole = dtf1['Job-Role'].value_counts(sort=True, ascending=False)
s2_jobrole = dtf2['Job-Role'].value_counts(sort=True, ascending=False)

#first bar plot with filter1
fig, ax1 = plt.subplots()
ax1.bar(s1_jobrole.index, s1_jobrole)
ax1.set_title("Individuals who earn < 50K have a High School Graduate Diploma and are in the United States")
ax1.tick_params(axis='x', labelrotation = 90)
#ax1.set_xticklabels(s1_jobrole.index, rotation=90)

#second bar plot with filter2
fig, ax2 = plt.subplots()
ax2.bar(s2_jobrole.index, s2_jobrole)
ax2.set_title("Individuals who earn > 50K have a High School Graduate Diploma and are in the United States")
ax2.tick_params(axis='x', labelrotation = 90)
#ax.set_xticklabels(s2_jobrole.index, rotation=90)

plt.show()

```

<b> Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn >50K? </b>

<b> Put your code below </b>

```python
#filter individuals who earn >50k
dtf3 = dtf_filter[ dtf_filter["Earnings"] == " >50K" ]

#to count the individuals of each Job Role, we subset the Job-Role column and use the value_counts method
#we use the sort argument to sort the Job Role (default True)
#we use the ascending argument to get the Job Role with the biggest count on top (default False)
#we also use the normalize argument to have proportion of the total (default False)
s3_jobrole = dtf3['Job-Role'].value_counts(normalize=True, sort=True, ascending=False)

display(s3_jobrole)

print(round(s3_jobrole[0],4)*100,'% of the individuals who earn >50K have', s3_jobrole.index[0], "Job Role.")
print('Job Role with the highest proportion of individuals who earn >50K: ', s3_jobrole.index[0], '(', (round(s3_jobrole[0],4)*100),'%', ')')
```
